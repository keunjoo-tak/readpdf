```python
# Smithers.pdf 들어있는 폴더에서 하나씩 불러다가 전처리후 txt로 저장

for i in pdf_list:
    pdf_path = f'./file/{i}'
    print('변환시작 :',pdf_path)
    parsed = parser.from_file(pdf_path)
    i_txt = open(f'./output/{i}.txt', 'w', encoding='utf-8')

    content_del_blank = parsed['content'][385:] # 표지부분 제거
    content_prepro = content_del_blank.replace('\n', ' ')
    content_prepro = content_prepro.replace(')  ', ')\n\n')
    content_prepro = re.sub("The Smithers Report - \w+ \w{2}, \w{4} Page \w+ of \w+", "", content_prepro)

    txt = ""
    final_sent = re.compile("[a-z]*\. ")

    for j in range(len(content_prepro)):
        txt = txt + content_prepro[j]
        m = final_sent.findall(content_prepro[j])
        if m:
            txt = txt + "\n"
    
    print(txt, file = i_txt)
    i_txt.close()

    # # 파파고 번역
    # txt_para_list = txt.split('\n')
    # txt_para_list = [v for v in txt_para_list if v]
    
    # for k in range(len(txt_para_list)-1) :
    #     params = {"source": "en", "target": "ko", "text": txt_para_list[k]}
    #     response =requests.post(request_url, headers=headers, data=params)
    #     result = response.json()
    #     print(result)
    #     # if result["message"] :
    #     print("원문 : ", txt_para_list[k], file = i_txt)
    #     print("번역결과 : ", result["message"]["result"]["translatedText"], file = i_txt)
    #     # else : 
    #         # pass
    # i_txt.close()            
```
